### 4.2. Illustration of the Proposed Method on the Synthetic Data

```python
model = 'MF' ; dataset = 'synthetic3'
baseline = pd.read_csv(f'../experiments/{model}_synthetic3_none_10_1.0_burninno_regno.csv')
pos2neg2 = pd.read_csv(f'../experiments/{model}_synthetic3_pos2neg2_10_0.8_burninno_regno.csv')
zerosum = pd.read_csv(f'../experiments/{model}_synthetic3_posneg_10_0.8_burninno_regno.csv')

result = pd.concat([baseline.tail(1), 
           pos2neg2.tail(1),
           zerosum.tail(1)
          ])
result['err']= (1-result['HR'])*100
result = result[['sample', 'HR', 'err', 'SCC_rank', 'mean']]
result.columns = ['method', 'accuracy', 'error', 'iPO', 'PopQ@1']
result['iPO'] = -result['iPO']
result['method'] = ['baseline', 'pos2neg2', 'zerosum']

result
```

### 4.3. Comparison with Earlier Methods

```python
model = 'MF'
ipw = pd.read_csv(f'../experiments/{model}_synthetic3_ipw_10_1.0_burninno_regno.csv')
PD = pd.read_csv(f'../experiments/{model}_synthetic3_pd_10_1.0_burninno_regno.csv')
pearson = pd.read_csv(f'../experiments/{model}_synthetic3_pearson_10_10.0_burninno_regno.csv')

result = pd.concat([ipw.tail(1), 
           PD.tail(1),
           pearson.tail(1)])
result['err']= (1-result['HR'])*100
result = result[['sample', 'HR', 'err', 'SCC_rank', 'mean']]
result.columns = ['method', 'accuracy', 'error', 'iPO', 'PopQ@1']
result['method'] = ['ipw', 'pd', 'pearson']
result
```
